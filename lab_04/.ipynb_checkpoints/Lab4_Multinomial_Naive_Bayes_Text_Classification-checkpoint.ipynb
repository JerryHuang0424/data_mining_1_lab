## Part 3. Text Preprocessing and Feature Extraction

### Task 1: Convert text into features

Machine learning models cannot use raw text directly.  
You will transform the documents into numerical vectors using:

1. **CountVectorizer** (bag-of-words)
2. **TfidfVectorizer** (term frequency-inverse document frequency)

### TODO
- Import `CountVectorizer` and `TfidfVectorizer`
- Create a count-based representation of the training and test data
- Create a TF-IDF representation of the training and test data
- Print the shapes of the resulting matrices

```python
# TODO: Import CountVectorizer and TfidfVectorizer

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
# TODO: Create bag-of-words features using CountVectorizer
# Suggested settings:
# stop_words="english", max_features=5000
vectorizer = CountVectorizer(stop_words="english", max_features=50000)

X_train_count = vectorizer.fit_transform(train_subset.data)
X_test_count = vectorizer.transform(test_subset.data)

# x= vectorizer.fit_transform(train_subset.data) is equal to:
# vectorizer.fit(train_subset.data)  # Learn the vocabulary from the training data
# x = vectorizer.transform(train_subset.data)  # Transform the training data into a matrix
#We create just one varible x, but vectorizer is updated into a new one.
```

```python
# TODO: Create TF-IDF features using TfidfVectorizer
# Suggested settings:
# stop_words="english", max_features=5000

tfidf_vectorizer = TfidfVectorizer(stop_words="english", max_features=50000)
X_train_tfidf = tfidf_vectorizer.fit_transform(train_subset.data)
X_test_tfidf = tfidf_vectorizer.transform(test_subset.data)
```

```python
# Print out result
print('CountVectorizer:')
print(f'Training set shape: {X_train_count.shape}')
print(f'Testing set shape: {X_test_count.shape}')

print('\nTifidfVectorizer:')
print(f'Training set shape: {X_train_tfidf.shape}')
print(f'Testing set shape: {X_test_tfidf.shape}')
      
```

**Why the feature matrix is sparse**

The feature matrix is sparse because each docmuent contains only a small subset of all possible words in the vocabulary. Since he vocavulary can include thousands of uniques word, most entries in the document-term matrix are zore. This results in a sprse representation where only a few elements per document have non-zore values.

### Student Notes
Answer the following:
1. What is a document-term matrix?
2. What is the difference between CountVectorizer and TfidfVectorizer?
3. Why do we remove stop words?
4. Why is `max_features` useful?

1. A document-term matrix is a numerical representation of a text dataset where each row corresponds to a document and each column corresponds to a unique term(word) from the vocabulary. The values in the matrix represent the frequency or importance(e.g., TF-IDF score) of each term in each document.
2. CountVectorizer converts text into a matrix of raw word counts, where each value represents how many times a wrod appears in a document. In contrast, TfidVectorizer computes the TF-IDF score, which reflects not only the frequency of a word in a document but also how important the word is across the entire corpus. TF-IDF reduces the influence of common words and emphasized more informative terms.
3. Stop words are common words such as "the", "is", and "and" that appear frequently but carry little meaningful information. Removing them helps reduce noise in the data, decreases dimensionality, and improves model performance by focusing on more informative words.
4. The max_features parameter limits the number of features (i.e., vocabulary size) used by the vectorizer. This helps reduce computational cost, prevents overfitting, and removes less important words by keeping only the most frequent or relevant terms.

## Part 4. Model Training and Evaluation

### Task 1: Train a Multinomial Naive Bayes classifier

You will train two models:
1. Multinomial Naive Bayes with CountVectorizer features
2. Multinomial Naive Bayes with TF-IDF features

### TODO
- Import `MultinomialNB`
- Train a model on count features
- Predict labels for the test set
- Compute the test accuracy

```python
# TODO: Import MultinomialNB and accuracy_score

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
# TODO: Train Multinomial Naive Bayes on count features
# TODO: Predict the test labels
# TODO: Print the accuracy

model_Multi = MultinomialNB()
model_Multi.fit(X_train_count, train_subset.target)

y_pred = model_Multi.predict(X_test_count)

accuracy = accuracy_score(test_subset.target, y_pred)

print(f'Accuracy of MultinomialNB with CountVectorizer: {accuracy:.4f}')
```

### Task 2: Evaluate the count-based model

### TODO
- Import and print a classification report
- Include precision, recall, and F1-score

```python
# TODO: Evaluate the count-based model with classification_report

from sklearn.metrics import classification_report

print('Classification Report(CounterVectorizer):')
print(classification_report(test_subset.target, y_pred, target_names=test_subset.target_names))
```

### Task 3: Train and evaluate the TF-IDF model

### TODO
- Train Multinomial Naive Bayes on TF-IDF features
- Predict the test labels
- Print the accuracy
- Print the classification report

```python
# TODO: Train and evaluate the TF-IDF model

tfidf_model = MultinomialNB()
tfidf_model.fit(X_train_tfidf, train_subset.target)

y_pred_tfidf = tfidf_model.predict(X_test_tfidf)
accuracy_tfidf = accuracy_score(test_subset.target, y_pred_tfidf)
print(f'Accuracy of MultinomialNB with TfidfVectorizer: {accuracy_tfidf:.4f}')
```

```python
print('Classification Report(TfidfVectorizer):')
print(classification_report(test_subset.target, y_pred_tfidf, target_names=test_subset.target_names))
```

### Task 4: Compare the two models

### TODO
- Create a summary table with:
  - model name
  - accuracy
  - weighted precision
  - weighted recall
  - weighted F1-score
- Decide which model performs better

```python
# TODO: Create a performance comparison table for the two models

from sklearn.metrics import accuracy_score, precision_recall_fscore_support
import pandas as pd

results = []

# ===== CountVectorizer 模型 =====
precision, recall, f1, _ = precision_recall_fscore_support(
    test_subset.target,
    y_pred,
    average='weighted'
)

results.append({
    "Model": "CountVectorizer + MultinomialNB",
    "Accuracy": accuracy_score(test_subset.target, y_pred),
    "Precision": precision,
    "Recall": recall,
    "F1-score": f1
})


# ===== TF-IDF 模型 =====
precision, recall, f1, _ = precision_recall_fscore_support(
    test_subset.target,
    y_pred_tfidf,
    average='weighted'
)

results.append({
    "Model": "TfidfVectorizer + MultinomialNB",
    "Accuracy": accuracy_score(test_subset.target, y_pred_tfidf),
    "Precision": precision,
    "Recall": recall,
    "F1-score": f1
})


# 转成 DataFrame
df_results = pd.DataFrame(results)

print(df_results)
```

### Task 5: Plot a confusion matrix

### TODO
- Import confusion matrix tools
- Plot the confusion matrix for the TF-IDF model
- Identify categories that are commonly confused

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

# 计算混淆矩阵
cm = confusion_matrix(test_subset.target, y_pred_tfidf)

# 归一化（按行）
cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

# 设置画布
plt.figure(figsize=(10, 8))

# 使用seaborn画 heatmap
sns.heatmap(
    cm_norm,
    cmap="Blues",
    xticklabels=test_subset.target_names,
    yticklabels=test_subset.target_names,
    annot=False,        
    fmt=".2f",
    cbar=True
)

# 只标注较大的值（提高可读性）
threshold = 0.2
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        if cm_norm[i, j] > threshold:
            plt.text(j + 0.5, i + 0.5,
                     f"{cm_norm[i, j]:.2f}",
                     ha="center", va="center",
                     color="black", fontsize=8)

# 美化
plt.xticks(rotation=60, ha='right', fontsize=9)
plt.yticks(rotation=0, fontsize=9)

plt.xlabel("Predicted Label", fontsize=12)
plt.ylabel("True Label", fontsize=12)
plt.title("Normalized Confusion Matrix (TF-IDF + MultinomialNB)", fontsize=14)

plt.tight_layout()
plt.show()
```

### Student Notes
Discuss:
- Which feature representation works better?
- Which categories are easiest to classify?
- Which categories are most often confused?
- Why do some categories overlap?

1. The TF-IDF feature representation performs better than the CountVectorizer representation. It achieves higher accuracy, precision, recall, and F1-score. This is because TF-IDF not only considers term frequency but also reduces the impact of commonly occurring words, allowing the model to focus on more informative and discriminative features.
   
2. Categories with highly distinctive vocabulary are the easiest to classify. For example, topics such as sci.space or rec.sport.hockey often contain domain-specific terms that make them easier for the model to identify correctly.

3. Categories within similar domains are most often confused. For instance, computer-related categories such as comp.graphics and comp.os.ms-windows.misc, or sports-related categories such as rec.sport.baseball and rec.sport.hockey, are frequently misclassified due to their similar vocabulary and context.

4. Some categories overlap because they share similar vocabulary and semantic context. Since bag-of-words models like CountVectorizer and TF-IDF ignore word order and deeper semantic meaning, they cannot fully distinguish between topics that use similar terms. As a result, documents from related categories may appear similar in the feature space, leading to misclassification.

Write a short conclusion that includes:
- how the text was transformed into features
- how Multinomial Naive Bayes was used
- whether CountVectorizer or TF-IDF performed better
- what the main limitations of the model are
- how you could improve the classifier in future work

In this project, raw text data was transformed into numerical features using both CountVectorizer and TF-IDF representations, which convert documents into a document-term matrix based on word frequencies and term importance. A Multinomial Naive Bayes classifier was then trained on these features to perform text classification by learning the probability distribution of words for each category.

Experimental results show that the TF-IDF representation outperforms the CountVectorizer approach, achieving higher accuracy and better overall performance. This is because TF-IDF reduces the impact of common words and emphasizes more informative terms.

However, the model has several limitations. The bag-of-words representation ignores word order and semantic context, and the Naive Bayes model assumes feature independence, which is often unrealistic. These limitations can lead to confusion between categories with similar vocabulary.

In future work, the classifier could be improved by using more advanced feature representations such as word embeddings, applying models that capture context (e.g., deep learning approaches), or tuning hyperparameters to optimize performance.
