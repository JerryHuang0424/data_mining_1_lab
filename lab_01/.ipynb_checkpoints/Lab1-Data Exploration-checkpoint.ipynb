## 5. Data Cleaning

```python
# Make a copy before cleaning
df_clean = df.copy()

# Remove duplicates
df_clean = df_clean.drop_duplicates()


df_clean = df_clean.drop(columns=['ID'])  # Assuming 'ID' is not useful for modeling

df_clean["Ever_Married"] = df_clean["Ever_Married"].fillna("No")  # Assuming 'No' is a reasonable default


for col in df_clean.select_dtypes(include='object').columns:
    if col != 'Ever_Married':  
        if df_clean[col].isnull().sum() > 0:
            df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])

for col in df_clean.select_dtypes(include=np.number).columns:
    if df_clean[col].isnull().sum() > 0:
        df_clean[col] = df_clean[col].fillna(df_clean[col].median())

print('Cleaned shape:', df_clean.shape)
df_clean.isnull().sum()
```

## 8. Encoding Categorical Variables

```python
df_encoded = df_clean.copy()

# Example: label encode object columns except the target if you want to keep it readable
label_encoders = {}
for col in df_encoded.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
    label_encoders[col] = le

df_encoded.head()
```

## 9. Feature Scaling

```python
df_scaled = df_encoded.copy()

scale_cols = df_scaled.select_dtypes(include=np.number).columns.tolist()

# TODO: Remove ID or target columns from scaling if needed
exclude_from_scaling = ["Segmentation"]  # Assuming 'Segmentation' is the target variable
scale_cols = [col for col in scale_cols if col not in exclude_from_scaling]

scaler = StandardScaler()
df_scaled[scale_cols] = scaler.fit_transform(df_scaled[scale_cols])

df_scaled.head()
```

### Preprocessing Pipeline Summary

**Missing Value Treatment:**
- **ID column**: Dropped entirely (24.56% missing, not useful for modeling)
- **Categorical variables** (Ever_Married, Graduated, Profession, Var_1): Imputed with mode (most frequent value)
- **Numerical variables** (Work_Experience, Family_Size): Imputed with median (robust to outliers)
- **Strategy rationale**: Preserves maximum data while maintaining distribution integrity

**Duplicate Handling:**
- **22 duplicate rows** identified and removed
- **Purpose**: Ensures each customer record is unique for accurate modeling

**Encoding Method:**
- **Label Encoding** applied to all categorical variables (Gender, Ever_Married, Graduated, Profession, Var_1, Spending_Score)
- **Target variable** (Segmentation) also label encoded for consistency
- **Advantage**: Maintains ordinal relationships where applicable while converting to numerical format

**Scaling Method:**
- **StandardScaler** applied to all numerical features
- **Target variable** (Segmentation) excluded from scaling
- **Purpose**: Normalizes feature scales to prevent dominance by features with larger ranges
- **Result**: All features have mean=0 and standard deviation=1

**Overall Pipeline Flow:**
1. Load raw data → 2. Remove duplicates → 3. Drop ID column → 4. Impute missing values → 5. Label encode categoricals → 6. Standard scale numericals → 7. Ready for modeling
