## 5. Data Cleaning

```python
# Make a copy before cleaning
df_clean = df.copy()

# Remove duplicates
df_clean = df_clean.drop_duplicates()

# Example imputation strategy
# TODO: Adjust based on your analysis
for col in df_clean.select_dtypes(include='object').columns:
    if df_clean[col].isnull().sum() > 0:
        df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])

for col in df_clean.select_dtypes(include=np.number).columns:
    if df_clean[col].isnull().sum() > 0:
        df_clean[col] = df_clean[col].fillna(df_clean[col].median())

print('Cleaned shape:', df_clean.shape)
df_clean.isnull().sum()
```

## 8. Encoding Categorical Variables

```python
df_encoded = df_clean.copy()

# Example: label encode object columns except the target if you want to keep it readable
label_encoders = {}
for col in df_encoded.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
    label_encoders[col] = le

df_encoded.head()
```

## 9. Feature Scaling

```python
df_scaled = df_encoded.copy()

scale_cols = df_scaled.select_dtypes(include=np.number).columns.tolist()

# TODO: Remove ID or target columns from scaling if needed
exclude_from_scaling = []
scale_cols = [col for col in scale_cols if col not in exclude_from_scaling]

scaler = StandardScaler()
df_scaled[scale_cols] = scaler.fit_transform(df_scaled[scale_cols])

df_scaled.head()
```
