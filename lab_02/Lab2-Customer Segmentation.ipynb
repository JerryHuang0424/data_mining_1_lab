```python
missing_count = df.isnull().sum() # 计算每列的缺失值数量
missing_percentage = ((missing_count / len(df)) * 100).round(2) # 计算每列的缺失值百分比

missing_table = pd.DataFrame(
    {
        'Missing Count': missing_count,
        'Missing Percentage (%)': missing_percentage
    }
).sort_values(by='Missing Percentage (%)', ascending=False) # 按照缺失值百分比降序排序

missing_table
```

## Comparison Task

Students must compare the three algorithms from the following aspects:

    (1) Predictive Performance
        Which model has the highest:
        accuracy
        F1-score
        balanced performance across all classes

    (2) Overfitting Risk
        Did the decision tree overfit?
        Did the neural network show unstable behavior?
        Was logistic regression more stable?

    (3) Interpretability
        Decision tree is usually easiest to explain
        Logistic regression provides linear relationships
        Neural network is often hardest to interpret

    (4) Computational Complexity
        Which model trained fastest?
        Which model required the most preprocessing?
        Which model was hardest to tune?

    (5) Suitability for This Dataset
        Students should discuss which algorithm is best for this specific dataset, not in general.
