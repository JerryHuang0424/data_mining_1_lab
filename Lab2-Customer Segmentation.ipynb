# Lab2: Multi-class Customer Segmentation

Target Variable: Segmentation

Algorithms:

    Decision Tree

    Logistic Regression

    Neural Network

## Experiment Goal

Students will build classification models to predict the customer segment of each individual using demographic and behavioral features. They will then compare model performance and explain which algorithm is most suitable for this dataset.

## Learning Objectives

1. By the end of the experiment, students should be able to:

1. Prepare data for classification tasks

1. Split data into training and testing sets

1. Train three different classifiers

1. Evaluate models using multiple metrics

1. Compare algorithm performance and interpret results

1. Explain advantages and disadvantages of each method

## Background of the Task

The dataset contains customer information such as:

    Gender

    Age

    Ever_Married

    Profession

    Work_Experience

    Spending_Score

    Family_Size

    Var_1

The prediction target is:

    Segmentation: A, B, C, or D

This is a multi-class classification problem.

## Experimental Requirements

Students must complete the following tasks:

### Part 1. Data Preparation

Students should:

    Load the dataset

    Inspect shape, column names, and data types

    Handle missing values

    Remove duplicates if necessary

    Encode categorical variables

    Scale features where appropriate

    Separate features X and target y

### Part 2. Train-Test Split

Students divide the dataset into:

    Training set: 80%

    Testing set: 20%

Use a fixed random seed such as random_state=42.

## Models to Be Built
### Model 1: Decision Tree

Students train a decision tree classifier.

Suggested settings:

    criterion="gini" or criterion="entropy"

    control tree depth to reduce overfitting

Students should observe:

    whether the tree overfits

    whether the model is easy to interpret

### Model 2: Logistic Regression

Students train a logistic regression classifier for multi-class prediction.

Suggested settings:

    use proper solver such as lbfgs

    increase max_iter if needed

    standardize features before training

Students should observe:

    whether linear decision boundaries are sufficient

    whether performance is stable

### Model 3: Neural Network

Students train a simple feedforward neural network.

Suggested settings:

    input layer = number of features

    1–2 hidden layers

    output layer = 4 classes

    activation: ReLU for hidden layers，softmax for output layer

Students may implement:

    MLPClassifier from scikit-learn or a Keras/TensorFlow model if allowed

Students should observe:

    whether the neural network improves performance

    whether training is more complex than the other methods

## Evaluation Metrics

Students must evaluate all three models using the same test set.

Required metrics:

    Accuracy

    Precision

    Recall

    F1-score

    Confusion Matrix

For multi-class classification, students should report:

    macro average

    weighted average

This helps them compare overall performance more fairly.

## Comparison Task

Students must compare the three algorithms from the following aspects:

    (1) Predictive Performance
        Which model has the highest:
        accuracy
        F1-score
        balanced performance across all classes

    (2) Overfitting Risk
        Did the decision tree overfit?
        Did the neural network show unstable behavior?
        Was logistic regression more stable?

    (3) Interpretability
        Decision tree is usually easiest to explain
        Logistic regression provides linear relationships
        Neural network is often hardest to interpret

    (4) Computational Complexity
        Which model trained fastest?
        Which model required the most preprocessing?
        Which model was hardest to tune?

    (5) Suitability for This Dataset
        Students should discuss which algorithm is best for this specific dataset, not in general.
